### **Barra baja**
| Notación     | Ejemplo    | Significado                                 |
|-----------------|--------------|-------------------------------------------|
| `attribute`     | `name`       | Público                                   |
| `_attribute`    | `_balance`   | Uso interno (convención)                  |
| `__attribute`   | `__password` | Name mangling (ocultación más fuerte)     |
| `attribute_`    | `class_`     | Evitar conflicto con palabras clave       |
| `__attribute__` | `__init__`   | Método especial de Python                 |




````python
# _variable means --> please, don't touch me directly, though you can

class Chair:
    _material = "Iron"
    
    def get_material(self):
        return self._material
    
one_chair = Chair()

one_chair.get_material() # because there is no instance attribute _material, it takes the attribute from the class
    
````

````python
# please, don't do this, the first underscore means "please do not touch the variable outside of the class, though you can"

Chair._material = "Wood" # I'm a rebel, I insert my hand inside the object and change the attribute although it has an initial underscore

one_chair.get_material()

# the initial underscore means: if you want to use it, use it through methods of the class, do not take it directly

````

````python
# one underscore in the end to avoid conflicts

if_ = "now I can use the name if"
if_
````

````python
# dunder (down under) for magic methods

class Table:
    
    def __init__(self, material): # dunder init
        self.material = material

my_table = Table("Stone")
my_table.material
````

````python
# double underscore in the beginning --> Python places the name of the class in the beginning
# This has to do with inheritance!!
````

### **Herencia**

El mecanismo de herencia de clases permite múltiples clases base; una clase derivada puede sobrescribir cualquier método de sus clases base, y un método puede llamar al método de una clase base con el mismo nombre.

Los objetos pueden contener cantidades y tipos arbitrarios de datos.

Al igual que ocurre con los módulos, las clases forman parte de la naturaleza dinámica de Python: se crean en tiempo de ejecución y pueden modificarse incluso después de haber sido creadas.



````python
class test:
    def __init__(self, name):
        self.__name = name # notice here the __name !!
        
    def __str__(self):
        return "my name is " + self.__name
        
t1 = test("Alberto")
vars(t1)
````

````python

# vars returns a dictionary with the data of the object
help(vars)
````

````python
# Observa cómo __name se ha almacenado en la clase como _test__name
````

````python
class other(test): # notice how other inherits from test !!
    
    def __init__(self, name, surname):  # it initialises with the name and surname 
        super().__init__(name)  # using super it sends the name to the dunder init of the upper class
        self.surname = surname  # the surname is in this class, in other
        
    def __str__(self):
        return super().__str__() + " and my surname is " + self.surname
        # I use the dunder str from the upper class, which returns the name and keep on going
        
o1 = other("Alberto", "Romero")

str(o1)
````
