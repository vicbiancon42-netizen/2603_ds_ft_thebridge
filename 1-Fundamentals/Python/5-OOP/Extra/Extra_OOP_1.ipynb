### Objetos como copias o como referencias

Cuando pasamos un objeto a un parámetro de una función, los objetos que son **colecciones** (lista, tupla, diccionario, conjunto) se pasan como **referencias**.<br>
Los objetos que no son colecciones se pasan como **copias**. 

Una **copia** es un gemelo del objeto original, pero **no** es el mismo objeto.<br>
Una **referencia** es, literalmente, el mismo objeto.

```python
def to_upper(name): # a str name, it is passed as a COPY  
    return name.upper() # the str name here is a copy from the argument, the name inserted in the argument is NOT changed

name = "Alberto"
returned_value = to_upper(name)

print('returned_value', returned_value)
print('name', name)
```

```python
def add_object(my_list, my_object): # a list passed as an argument is passed DIRECTLY (a reference to the object)
    my_list.append(my_object) # the same list that appears in the argument, it can be changed!!!
    return my_list

my_list = []
my_object = 2
returned_object = add_object(my_list, my_object)

print('returned_value', returned_value)
print('my_list', my_list)
```

### Ejemplo de una clase

```python
class RollerCoaster:
    
    material = "iron"
    prize_list = []
    
    def __init__(self, name = "Tornado", inspection_list = []):
        self.name = name
        self.inspection_list = inspection_list
        
    def add_prize(self, prize):
        self.prize_list.append(prize)
        
    def add_inspection(self, inspection_year):
        self.inspection_list.append(inspection_year)
        

# I initiate a roller coaster
my_roller_coaster = RollerCoaster()

print("material", my_roller_coaster.material) # I access the class attribute

my_roller_coaster.add_prize("best attraction")
my_roller_coaster.add_inspection(2025)

print("prize list", my_roller_coaster.prize_list)
print("inspection list", my_roller_coaster.inspection_list)
```

Fíjate en los **atributos de clase**. <br> Todas las instancias de esa clase tendrán esos atributos. Si el atributo de la clase es mutable, todas las instancias de la clase compartirán el mismo objeto como atributo.<br> Si el atributo de la clase es inmutable, no podemos cambiarlo. Si lo intentamos, la instancia tendrá su propio atributo con su propio valor, pero la instancia no cambiará el atributo de clase.

Observa los **atributos de instancia** en el método __init__. <br> Cada instancia tendrá esos atributos.

Ojo con **self**.<br>**No** es una palabra reservada de Python, solo es el nombre de una variable (pero es buena idea llamarla *self* porque otros desarrolladores entenderán lo que queremos decir).<br>*self* se refiere a la instancia a la que se pueden aplicar los métodos de instancia.

```python
# I initiate another roller coaster
my_roller_coaster2 = RollerCoaster("twister", [])

print("material", my_roller_coaster2.material) # I access the class attribute

my_roller_coaster2.add_prize("amazing attraction")
my_roller_coaster2.add_inspection(2024)

print("prize list", my_roller_coaster2.prize_list) # I access the class attribute
print("inspection list", my_roller_coaster2.inspection_list) # I access the instance attribute
```

Podemos acceder a los atributos de la clase a través de una instancia o a través de la clase:

```python
RollerCoaster.prize_list
```

```python
my_roller_coaster.prize_list
```

Podemos acceder a los atributos de la instancia a través de la instancia:

```python
my_roller_coaster.inspection_list
```

```python
my_roller_coaster2.inspection_list
```

Los atributos de clase se comparten entre todas las instancias de la clase.
Si el atributo de clase es mutable, cualquier instancia puede cambiarlo.
Si el atributo de clase es inmutable, no se puede cambiar. Si una instancia intenta cambiarlo, esa instancia tendrá su propia versión del atributo.

```python
my_roller_coaster2.material = "wood" 
```

```python
my_roller_coaster.material
```

```python
my_roller_coaster2.material
```
