**Métodos y propiedades privadas**

En Python no existen métodos o propiedades verdaderamente privadas: todo es accesible desde fuera de la clase. Sin embargo, por convención puedes anteponer un guion bajo a métodos y propiedades para indicar que deben tratarse como privados (por ejemplo, `_private_method`). Esto es solo una señal para otros desarrolladores de que son componentes internos que no deberían usarse desde el exterior.

**¿Qué son los métodos privados?**

Los métodos privados en la programación orientada a objetos (POO) son funciones diseñadas para usarse únicamente dentro de la propia clase. Ayudan a gestionar la funcionalidad interna de una clase, protegiéndola del acceso directo externo. Sin embargo, en Python aún puedes acceder a estos métodos si lo deseas.

Ahora que ya comprendes lo básico de la POO en Python, podemos pasar a los *magic methods*, métodos especiales que permiten realizar operaciones sofisticadas con objetos.

### Implementación de Magic Methods

Los *magic methods* son métodos especiales predefinidos en Python que se reconocen fácilmente por los dobles guiones bajos al inicio y al final de sus nombres (por ejemplo, `__init__`). Estos métodos no los llamas directamente: el intérprete de Python los utiliza cuando trabaja con distintos elementos como operadores o funciones integradas.

Para entender su uso práctico, exploremos algunos de los más comunes, empezando por `__repr__` y `__str__`.

**Representación de objetos: `__repr__` y `__str__`**

Al crear una clase, suele ser útil definir cómo debe representarse una instancia de esa clase como cadena. Python ofrece dos *magic methods* para este propósito: `__repr__` y `__str__`. Veamos más de cerca cómo funciona cada uno.

```python
class Distance:        
    def __init__(self, value, scale):
                self.value = value
                self.scale = scale
    def __repr__(self):
            return f"Distance --> ({self.value}, {self.scale!r})"  # !r add quotes
    def __str__(self):
            return f"Distance is {self.value} -- {self.scale}"
        
d = Distance(10, "Km")

print(repr(d))    # "Distance(10, 'Km')"
print(str(d))    # "Distance is 10 Km"
print(d)
```

**Métodos de comparación – `__eq__`, `__gt__`, `__lt__`, y otros**

Comparar dos distancias con unidades distintas produciría resultados inesperados. Los *magic methods* permiten sobrecargar los operadores por defecto para realizar comparaciones con sentido. A partir del ejemplo anterior, podemos ampliarlo para incorporar este tipo de comportamiento.

```python
class Distance:
    def __init__(self, value, scale): # magic or dunder method, because the underscores
        self.value = value
        self.scale = scale
        if scale == "Km":
            self.value_miles = value * 0.62
            self.value_km = value
        elif scale == "miles":
            self.value_miles = value
            self.value_km = value / 0.62
            
    def __eq__(self, other): # for ==
        return self.value_km == other.value_km
    def __lt__(self, other): # for less than, <
        return self.value_km < other.value_km
```

```python
Distance(10,"Km").value_miles # 10 Km are 6.2 miles
```

```python
d_km = Distance(10, "Km")
d_miles = Distance(6.2, "miles")
d_miles2 = Distance(12, "miles")
print(d_km == d_miles)    # True
print(d_km < d_miles2)    # True
```

**Eso es todo.** Si quieres disponer de todos los operadores de comparación, también deberías implementar los demás *magic methods* correspondientes: `__le__`, `__gt__` y `__ge__`, que cubren *menor o igual que*, *mayor que* y *mayor o igual que*, entre otros.

### Objetos callable  – `__call__`


El último método mágico que revisaremos es __call__. Este es un poco especial porque te permite llamar a una instancia de tu objeto como si fuera una función normal. Veamos un ejemplo:

```python
class Counter:
    def __init__(self):
        self.counter = 0
    def __call__(self, inc=1):
        self.counter += inc
        
my_counter = Counter()
print(my_counter.counter)    # 0
my_counter() # execute its call method!
print(my_counter.counter)    # 1
my_counter()
print(my_counter.counter)    # 2
```

La clase Waiter:<br><br>
La mayoría de camareros trabajan para el mismo restaurante, llamado "Beautiful restaurant".<br>

Las propinas se guardan en un diccionario cuyas claves son cadenas que contienen fechas ('17-01-2026') y cuyos valores son los euros (int).<br>
En un mismo día puede haber muchas propinas, así que usa una lista en los valores del diccionario.
<br>

Cada camarero tiene un id y un turno (mañana o tarde).
Además, si una propina es mayor que 100 €, la propina no se comparte y ese camarero la guarda en su propio diccionario great_tips. 

El restaurante abre otra ubicación donde trabajarán muy pocos camareros.<br>
Debes tener un método de instancia que permita guardar por defecto el nombre "Beautiful restaurant new boulevard" en la instancia.<br>

Debe haber un método de instancia que devuelva una lista con todas las posibles interacciones que puedes hacer con esa instancia.<br>

<br>

Debes tener un método de clase *empty_shared_tips* que devuelva la cantidad total de dinero compartido y establezca shared_tips como un diccionario vacío.<br>
Debes tener un método estático que imprima "We are really mastering objects with this exercise of waiters!"

Hay diferentes maneras de resolver este ejercicio.

Estas indicaciones están pensadas para que uses atributos de clase (mutables e inmutables), atributos de instancia (mutables e inmutables), métodos de clase, métodos estáticos y decoradores.<br>

```python
# La solución del profe :-)


class Waiter:
    
    # class attributes
    restaurant = "Beautiful restaurant"
    shared_tips = {}
    
    # Instance methods
    
    def __init__(self, id=0, shift="Morning"):
        self.id = id
        self.shift = shift
        self.great_tips = {}
        
    def to_share_tips(self, tip=0, day="01-01-2026"):
        if day in self.shared_tips:
            self.shared_tips[day].append(tip)
        else:
            self.shared_tips[day] = [tip]
            
    def to_great_tips(self, tip=100, day="01-01-2026"):
        if day in self.great_tips:
            self.great_tips[day].append(tip)
        else:
            self.great_tips[day] = [tip]
        
    def tip(self, tip=0, day="01-01-2026"):
        if tip <= 100:
            self.to_share_tips(tip, day)
        else:
            self.to_great_tips(tip, day)
            
    def restaurant_name(self, name="Beautiful restaurant new boulevard"):
        self.restaurant = name
        
    def interactions(self):
        return dir(self)
    
    # class method
            
    @classmethod
    def empty_shared_tips(cls):
        total = 0
        for day in cls.shared_tips:
            total += sum(cls.shared_tips[day])
        cls.shared_tips = {}
        return total
    
    # static method
    
    @staticmethod
    def print_learning_Python(): # being static, it won't consider the first parameter as the idea of self for instances or cls for the class
        print("We are really mastering objects with this exercise of waiters!")
    
# We create 3 waiters 
            
waiter1 = Waiter(id=0, shift="Evening")
waiter2 = Waiter(id=1, shift="Evening")
waiter3 = Waiter(id=2, shift="Morning")

# We tip them

waiter1.tip(tip=50, day="17-01-2026")
waiter2.tip(tip=75, day="17-01-2026")
waiter1.tip(tip=80, day= "18-01-2026")
waiter2.tip(tip=200,day="17-01-2026")

# We move waiter1 to the new restaurant location

waiter1.restaurant_name()

print ("waiter1 id:", waiter1.id, "works in", waiter1.restaurant)
print ("waiter2 id:", waiter2.id, "works in", waiter2.restaurant)

# We take a look at the shared tips
print("shared tips:", Waiter.shared_tips)
print()

# We take a look at waiter2 great tips 
print("great tips from waiter2:", waiter2.great_tips)
print()

# We take a look at all the interactions that we can do with waiter1
print("intercations with waiter1:", waiter1.interactions())

# We empty the shared tips
total_shared_tips = waiter1.empty_shared_tips()
print("total_shared_tips:", total_shared_tips)
print()

# We print the current shared tips
print("shared tips after emptying the can", Waiter.shared_tips)
print()

# We use a static method
Waiter.print_learning_Python()
    
```
