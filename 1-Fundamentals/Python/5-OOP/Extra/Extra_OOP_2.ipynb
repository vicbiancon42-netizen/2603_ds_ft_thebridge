### Ámbito de los objetos

Cada objeto tiene un alcance (¿desde dónde es visible?).

````python
# inner functions
# in Python every object has a scope: who can see it?
# the scopes are: global, local, non local
# now we are only interested in showing that functions (that are objects) can be inside functions (that are objects)

def parent():
    print("Printing from parent()")

    def first_child():
        print("Printing from first_child()")

    def second_child():
        print("Printing from second_child()")

    second_child()
    first_child()
````

````python
parent()
````

````python
def parent2(num):
    def first_child():
        return "Hi, I'm Alberto"

    def second_child():
        return "Call me Mr Romero"

    if num == 1:
        return first_child
    else:
        return second_child
````

````python
first = parent2(1)
second = parent2(2)
````

````python
first()
````

````python
second()
````

Bien, entonces las funciones son objetos. Veamos más ejemplos:

````python
def fun(a,b):
    return a+b

def function_that_uses_another_fun(one_func, a, b):
    return one_func(a,b)

function_that_uses_another_fun(fun,2,3)
````

En este ejemplo, estamos pasando una función y dos números (tres objetos) a una función (otro objeto), para devolver un objeto.
<br>
Analicemos otro ejemplo con objetos con diferentes ámbitos.

````python
# The scope of an object can be global, local or non local
# I can nest functions

colour = "green"  # global scope

def adding(a,b,c,d,e,f):
    name = "Alberto" # local scope
    
    def inner_adding(a,b,c,d,e):
        name = "Professor" # non local scope
        
        def inner_inner_adding(a,b,c,d):
            print(name, "level 2", colour)
            return a+b+c+d
        
        return e + inner_inner_adding(a,b,c,d)
    

    print(name, "level 1", colour)
    return f + inner_adding(a,b,c,d,e)


print(adding(1,1,1,1,1,1))
print(colour)
````

### Decoradores

El uso de un decorador es modificar el comportamiento predeterminado de una función.

````python
def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")
    return wrapper

def say_whee():
    print("Whee!")

say_whee = my_decorator(say_whee) # a decorator MODIFIES the default behaviour of say_whee

# The use of a decorator is to modify the default behaviour of a function
````

````python
say_whee
````

````python
say_whee() # we execute it
````

#### Un decorador envuelve una función, modificando su comportamiento.

````python
def another_decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")
    return wrapper

@another_decorator
def say_whee():
    print("Whee!")
````

````python
say_whee()
````

Ese @ se llama **pie syntax** y aparecerá mucho.
Sin embargo, ahora sabemos que esos @ son solo decoradores, que envuelven y modifican funciones :-)

### Generadores

Una función siempre devuelve un objeto.

````python
def my_function(num1, num2, num3):
    return [num1, num2, num3]

returned_value = my_function(1,2,3)

print("my returned value", returned_value)
````

````python
# Generators

# Generators do not return objects, they yield objects
# It's like a function BUT I yield the objects little by little, not altogether

def generator():
    yield 1
    yield 2
    
    for x in range(5):
        yield x
````

###  Comprehension y generadores

Se podría suponer que sustituir los corchetes por paréntesis crearía una tupla, pero en realidad se obtiene un objeto generador. La diferencia clave entre los generadores y las comprensiones de listas es que los generadores generan elementos sobre la marcha, en lugar de calcularlos y almacenarlos todos a la vez en la memoria. Se puede considerar un generador como un conjunto de instrucciones para producir valores.

Para definir un generador, simplemente se utiliza la misma sintaxis que una comprensión de lista, pero con paréntesis.

````python
(x**2 for x in [1,2,3,4,5])
````

Podemos hacer que el generador sea un poco más complejo, ¿puedes adivinar qué producirá este generador?

````python
(x**2 if x > 3 else 0 for x in [1,2,3,4,5])
````

Vamos a averiguarlo:

````python
my_generator = (x**2 if x > 3 else 0 for x in [1,2,3,4,5])

for element in my_generator:
    print(element)
````

#### Definición de una clase
Crear una clase en Python es fácil. Se empieza con la palabra clave `class`, seguida del nombre de la clase y, a continuación, un nuevo bloque sangrado donde se pueden definir los métodos. Es similar a cómo se definirían las funciones. A continuación se muestra un ejemplo básico para ilustrarlo.

````python
class RollerCoaster:
    def explain(self, name):
        return f"The name of this roller coaster is {name}"

my_roller_coaster = RollerCoaster()
print(my_roller_coaster.explain("Tornado"))    # "The name of this roller coaster is Tornado"
````

### `__init__`

Inicializar objetos (no, en realidad no es construir).

##### Volvamos a nuestro ejemplo de la montaña rusa.

````python
class RollerCoaster:
    '''
    This is a roller coaster class
    :Authors:
        Alberto Romero  
    '''
    
    def __init__(self, name):
        self.name = name
    
    def greet(self, passanger = "Anonymous"):
        return f"Hello, {passanger}, you are going to ride on the roller coast {self.name}"

my_roller_coaster = RollerCoaster("Rocky Mountains")
print(my_roller_coaster.greet("Alberto")) 
print(my_roller_coaster.name)
print(my_roller_coaster.greet())
````

En este ejemplo, el método `__init__` se utiliza para establecer una propiedad predeterminada para la montaña rusa: el nombre del pasajero. Si no se proporciona ningún nombre en el argumento, el método `greet` utilizará este nombre predeterminado. Como se muestra, se puede acceder fácilmente a esta propiedad utilizando la notación de puntos.

Sin embargo, es importante señalar que `__init__` **técnicamente no es un constructor**. En la mayoría de los lenguajes orientados a objetos, un constructor se encarga de crear un objeto en la memoria. En Python, el objeto ya está creado en la memoria cuando se llama a `__init__` (ya que tenemos acceso a la instancia `self`). Aunque Python tiene un método `__new__` que define la creación de objetos, rara vez se utiliza en los programas típicos de Python.
