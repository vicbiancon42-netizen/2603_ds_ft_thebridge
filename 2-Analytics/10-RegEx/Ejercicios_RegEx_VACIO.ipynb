## Ejercicio 1
Calcula la proporción de visitantes los dominogos del siguiente dataset.
Pasos recomendados:
- Entender el tipo de los datos
- Transformar el tipo de las fechas de los datos
- Crear una columna nueva llamada weekday

````python
import pandas as pd

# Run this cell to create the dataset
from datetime import datetime

dates = [f"2026-01-{day:02d}" for day in range(1, 32)]
visitors = [270, 470, 300, 420, 390, 360, 330, 300, 420, 490, 480, 410, 460, 480, 220, 450, 390, 430, 400, 350, 380, 290, 330, 490, 420, 470, 340, 500, 260, 420, 400]

df_visitations = pd.DataFrame({'date': dates, 'visitors': visitors})
df_visitations.head()
````

````python
df_visitations['date'] = pd.to_datetime(df_visitations['date'])
````

````python
df_visitations['weekday'] = df_visitations['date'].dt.day
df_visitations['weekday_name'] = df_visitations['date'].dt.day_name()
````

````python
df_visitations['date'].dt.strftime("%A")
````

````python
df_visitations['date'].dt.strftime("The date is %A, %d, the year is %Y")
````
