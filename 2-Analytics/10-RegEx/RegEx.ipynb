## 10.1 Patrones de expresiones regulares y metacaracteres
- Special characters:
  - `.`  - Any Character Except New Line
  - `\d` - digits (0-9)
  - `\D` - Not a Digit (0-9)
  - `\w` - word characters (a-z, A-Z, 0-9, _)
  - `\W` - Not word characters (a-z, A-Z, 0-9, _)
  - `\s` - whitespace (space, tab, newline)
  - `\S` - Not whitespace (space, tab, newline)
- Quantifiers:
  - `*` - 0 or more
  - `+` - 1 or more
  - `?` - 0 or 1
  - `{n}` - exact number
  - `{3,4}`   - Range of Numbers (Minimum 3, Maximum 4)
- Character sets `[]` and boundaries (`^`, `$`, `\b`)

  - `[]`  - Matches Characters in brackets
  - `[^ ]`    - Matches Characters NOT in brackets
  - `|`       - Either Or
  - `( )`     - Group

- Anchors
  - `\b`      - Word Boundary
  - `\B`      - Not a Word Boundary
  - `^`       - Beginning of a String
  - `$`       - End of a String 

- Capturing Groups *(allow you to group and extract parts of the match)*:
  - `(...)` - Creates a capturing group
- Lookarounds *(allow you to assert what comes before/after a match without including it in the match itself)*:
  - `(?=...)` - Positive lookahead
  - `(?!...)` - Negative lookahead
  - `(?<=...)` - Positive lookbehind
  - `(?<!...)` - Negative lookbehind


## Key Python Regex Methods
- `re.search()` - Finds first occurrence
- `re.match()` - Matches at beginning of string
- `re.findall()` - Finds all occurrences
- `re.finditer()` - Returns iterator of matches

### 10.5 Cuantificadores

```python
word = "Hello"

pattern = re.compile(r"Hello.?")  # 0 or 1
matches = pattern.finditer(word)
for match in matches:
    print(match)
```

```python
word = "Hello"

pattern = re.compile(r"Hello.*")  # 0 or more
matches = pattern.finditer(word)
for match in matches:
    print(match)
```

```python
word = "Hello"

pattern = re.compile(r"Hello.+")  # 1 or more in this case, there is no match
matches = pattern.finditer(word)
for match in matches:
    print(match)
```

### 10.9 Trabajando con logs

```python
with open('data_session10.txt') as f:
    content = f.read()

    print(content)
```

Open the file and filter out all emails, dates, number of like and text.

```python
import re

with open('./data_session10.txt') as f:
    content = f.read()

    email_pattern = re.compile(r'[a-zA-Z0-9]+@\w+\.\w+')
    emails = email_pattern.findall(content)

    date_pattern = re.compile(r'\d{4}-\d{1,2}-\d{1,2}')
    dates = date_pattern.findall(content)

    likes_pattern = re.compile(r'(?<=likes:)\d+') # positive lookbehind
    likes = likes_pattern.findall(content)

    reposts_pattern = re.compile(r'(?<=reposts:)\d+')
    reposts = reposts_pattern.findall(content)

    texts_pattern = re.compile(r'(?<=>\s).+(?=\n)') # positive lookahead
    text = texts_pattern.findall(content)
    
    # lookbehinds and lookaheads can be negative if you change = for !
```

Crea un dataframe de pandas que contenga los datos que has extraído.

```python
import pandas as pd
df = pd.DataFrame(
    {
        'email': emails,
        'dates': dates,
        'likes': likes,
        'reposts': reposts,
        'tweet': text
    }
)

df
```

```python
from bs4 import BeautifulSoup as bs
import requests
import time
```

```python
url = 'https://boardgamegeekstore.com/'
headers = {"User-Agent":"Mozilla/5.0"}
response = requests.get(url, headers=headers)
soup = bs(response.text, 'lxml')
```

```python
info_lista = soup.find_all('div', class_='grid-item__content')
for i in info_lista:
    print(i.get_text(strip=True))
    print(re.compile(r'(\$\d*\.+\d*)').findall(str(i)))
```
