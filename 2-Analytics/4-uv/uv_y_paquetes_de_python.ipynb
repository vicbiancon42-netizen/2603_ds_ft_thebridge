# UV
UV es un gestor de paquetes y proyectos de Python extremadamente rápido.

## Conceptos:
### Modulo
Un módulo es un objeto que sirve como una unidad organizativa del código Python. Los módulos tienen un namespace que contiene objetos Python arbitrarios. Los módulos se cargan en Python mediante el proceso de importing. Un namespace de Python es una asignación de nombres a objetos. Funciona como un diccionario donde las claves son nombres de objetos y los valores son los propios objetos.

### Paquete
Un paquete es un módulo de Python que puede contener submódulos o, recursivamente, subpaquetes. Técnicamente, un paquete es un módulo de Python con un atributo __path__. El atributo __path__ de un paquete se usa durante las importaciones de sus subpaquetes. Dentro del mecanismo de importación, funciona de manera muy similar a sys.path, es decir, proporcionando una lista de ubicaciones donde buscar módulos durante la importación.

- Los módulos son unidades de código.
- Los paquetes son unidades de entrega.

El ecosistema de Python no define qué es una librería.
Sin embargo, los desarrolladores tienden a llamar librerías a paquetes grandes con muchas herramientas.

Python define qué es la biblioteca estándar: la colección de paquetes, módulos y módulos de extensión distribuidos como parte del paquete oficial del intérprete de Python.

La guía de estilo de Python se apoya en los Python Enhancement Proposals, PEP.
No se siguen siempre todos los PEP, - por eso se llaman "Proposals".

Los módulos se nombran por los nombres de archivo, y los paquetes por el nombre del directorio.
En otras palabras, los nombres de archivo deben estar en minúsculas, con guiones bajos si eso mejora la legibilidad. De manera similar, los nombres de directorio deben estar en minúsculas, sin guiones bajos si es posible evitarlos.

Cualquier archivo Python (.py) es un módulo, y un montón de módulos en un directorio es un paquete. 
Para que un directorio sea un paquete debe tener un fichero  __init__.py

### Instalando paquetes en el entorno virtual

Queremos usar el paquete de python **ydata_profiling** pero no lo tenemos instalado previamente.



Ve a la carpeta del proyecto y escribe:

```bash
uv add ydata_profiling
```



Si vuelves a project-alberto-example, ahora hay un **.venv** (un entorno virtual llamado **.venv**, es una carpeta oculta) y un fichero **uv.lock**.



**uv.lock** es un archivo de bloqueo multiplataforma que contiene información exacta sobre las dependencias de tu proyecto.

A diferencia de **pyproject.toml**, que se utiliza para especificar los requisitos generales del proyecto, el archivo de bloqueo contiene las versiones exactas ya resueltas que **están instaladas en el entorno del proyecto**.

Este archivo debe incluirse en el **control de versiones, lo que permite instalaciones coherentes y reproducibles entre distintas máquinas**.

**uv.lock** es un archivo TOML legible para humanos, pero **está gestionado por uv y no debe editarse manualmente**.

En la carpeta **.venv** podemos comprobar que dentro de **Lib\site-packages** ahora tenemos **ydata_profiling**.



Añade setuptools, escribe 
```bash
uv add setuptools
```



Y ahora ejecuta el fichero **main.py**

## Creando una librería con python

Adicionalmente a las aplicaciones, **uv** puede crear paquetes y librerías.

Crea otro proyecto que se llame **project-alberto-library**.

```bash
uv init - - lib project-alberto-library
```





El fichero **README.md** es un fichero que contiene la introducción al proyecto (es el que se muestra por defecto cuando entras en un repo de GitHub).

El fichero **pyproject.toml** contiene la siguiente información:



Dentro de la carpeta **src** tenemos **project_alberto_library**.

La carpeta externa corresponde al proyecto.

Existe porque quieres que todo lo relacionado con un proyecto esté en la misma ubicación —especialmente para el **control de versiones**— y lleva **el nombre del proyecto**.

La carpeta interna corresponde a **la librería**.

Existe porque quieres que todo lo relacionado con **la aplicación esté agrupado, pero excluyendo archivos que son solo del proyecto —especialmente para el empaquetado—**.
Tienen el mismo nombre, pero NO se refieren al **mismo concepto**.

Como se trata de una librería, tenemos py.typed.

Si fuera un package, ese archivo no estaría ahí.

En el **dunder init** de la librería tenemos lo que se ejecutará cuando **importemos la librería**.
